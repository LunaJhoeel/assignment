### Task 1. Please find the total Valid Bet Amount in SGD?


First, lest check if there are any bets in the same target currency (SGD is the target)

```python
df_second_sheet['Currency'].value_counts()
```

There is no SGD currency, therefore all records must be converted

```python
# Convert the stake amount with the rate
df_second_sheet['AmountInSGD'] = df_second_sheet['Actual_Stake'] * df_second_sheet['ActualRate']

# Get the total sum with two decimal points
df_second_sheet['AmountInSGD'].sum().round(2)
```

Answer: The total valid amount in SGD is 50,162,771.37

### Task 2. From your observations of the data, please share any insights that you have.


The data corresponds to bets from a different Asian countries. From the metadata, the transactions belong to a website or online platform.

```python
# Set option to display large numbers
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Groups by currency and sums by total SGD and order by value
df_second_sheet.groupby('Currency')['AmountInSGD'].sum().reset_index().sort_values(by='AmountInSGD', ascending=False)
```

From the output, the top 3 countries that bet more ammount of money are Indonesia, THailand, and Vietnam.

```python
# Convert the win/lost amount with the rate
df_second_sheet['AmmountWinlost'] = df_second_sheet['Winlost'] * df_second_sheet['ActualRate']
# Groups by currency and sums by total SGD and order by value
df_second_sheet.groupby('Currency')['Winlost'].sum().reset_index().sort_values(by='Winlost', ascending=True)
```

From the output, the top 3 earner countries are KOrea, other countries, and Thailand


```python
# COunt the number of transactions
df_second_sheet.groupby('Currency')['Transaction Id'].count().reset_index().sort_values(by='Transaction Id', ascending=False)
```

From the output, the top 3 betting countries by transactions are Indonesia THailand, and Vietnam
